```python
#importing files, libraries, and modules and defining dataframes
import pandas as pd
import numpy as np
artists_df = pd.read_csv('artists.csv')
artworks_df = pd.read_csv('artworks.csv')
art_on_view_df = pd.read_excel('MoMA_OnView.xlsx')
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
print(artists_df.shape)
artists_df.head(2)
```

```python
print(artworks_df.shape)
artworks_df.head(2)
```

```python
print(art_on_view_df.shape)
art_on_view_df.head(2)
```

```python
# Converting ConstituentID to string type for consistency across dataframes
artists_df['ConstituentID'] = artists_df['ConstituentID'].astype(str)
artworks_df['ConstituentID'] = artworks_df['ConstituentID'].astype(str)
art_on_view_df['ConstituentID'] = art_on_view_df['ConstituentID'].astype(str)
```

```python
#Mergeing the dataframes
merged_df = artworks_df.merge(artists_df, on='ConstituentID', how='left')
print(merged_df.shape)
merged_df.head(2)
```

```python
merged_on_view_df = merged_df.merge(art_on_view_df, on='ConstituentID', how='left')
print(merged_on_view_df.shape)
merged_on_view_df.head(2)
```

```python
#Converting the 'Date' column to datetime format
merged_on_view_df['DateAcquired'] = pd.to_datetime(merged_on_view_df['DateAcquired'], errors='coerce')
```

```python
merged_on_view_df['Artist'].value_counts().head(5)
```

```python
#Main Analysis
num_unique_artists = merged_on_view_df['DisplayName_x'].nunique() #number of unique artists
top_nacionalities = merged_on_view_df['Nationality_y'].value_counts().head(10) #top 10 nationalities
gender_counts = merged_on_view_df['Gender_y'].value_counts() #gender counts
top_clasifications = merged_on_view_df['Classification'].value_counts().head(10) #top 10 classified collections
department_counts = merged_on_view_df['Department'].value_counts() #Artwor perdepartments
adquisition_trends = merged_on_view_df['DateAcquired'].dt.year.value_counts().sort_index() #Artworks acquired per year
top_artists = merged_on_view_df['DisplayName_y'].value_counts().head(10) #top 10 artists
num_on_display = merged_on_view_df['OnView'].notna().sum() #number of artworks on display
```

```python
import plotly.express as px

#Since plotly is not properly installed in the environment, we will use the default renderer to display the plot in a browser.
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
#Visualization of Top 10 Nationalities
fig_nationalities = px.bar(x=top_nacionalities.index,y=top_nacionalities.values,                             
                           title= 'Top 10 Nationalities of Artists in MoMA',
                           labels={'x':'Nationality', 'y':'Number of Artists'},
                           color=top_nacionalities.values)
fig_nationalities.show()
```

```python
#Visualization of Artworks by Genders

fig_gender = px.bar(x=gender_counts.index, y=gender_counts.values,
                    title='Gender Distribution of Artists in MoMA',
                    labels={'x':'Gender', 'y':'Number of Artists'},
                    color=gender_counts.values)
fig_gender.show()
```

```python
#Visualization of Artworks per Classification
fig_classification = px.bar(x=top_clasifications.index, y=top_clasifications.values,
                            title='Top 10 Classifications of Artworks in MoMA',
                            labels={'x':'Classification', 'y':'Number of Artworks'},
                            color=top_clasifications.values)
fig_classification.show()
```

```python
#Visualization of Artworks Acquired per Year
fig_acquisition = px.line(x=adquisition_trends.index, y=adquisition_trends.values,
                          title='Artworks Acquired per Year',
                          labels={'x':'Year', 'y':'Number of Artworks'},
                          markers=True,)
fig_acquisition.show()
```
